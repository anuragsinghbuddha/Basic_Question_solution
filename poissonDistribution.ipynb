## If on an average 6 person arrive bank in every 2 minute in bank in busy schedule , (a)what probability that exactly 4 customer arrive the bank in given time(b) what is probabity that more than  three customer arrive the bank in given time.

```python
import numpy as np 
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
rate=6
```

```python
n=np.arange(0,20)
```

```python
poisson=stats.poisson.pmf(n,rate)
poisson
```

```python
plt.plot(n,poisson,"o-")
plt.title('Poisson Distribution',fontsize=15)
plt.xlabel("Number of customer arriving in minute")
plt.ylabel("Probability of customer arriving in minute")
plt.show()
```
