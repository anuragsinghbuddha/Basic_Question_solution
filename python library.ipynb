#### Matplot is python library used for data visualisation

```python
import numpy as np
from  matplotlib import pyplot as plt
x=np.array([1,2,3,4,5,6])
y=(2*x)
p=plt.plot(x,y)
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
plt.plot(x,y)
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.title("Title")
```

```python
import numpy as np
from matplotlib import pyplot as plt
plt.plot(x,y,color ='r',linestyle=':',linewidth=5)
plt.xlabel("x-label")
plt.ylabel("y-label")
plt.title("Title")
```

```python
x=np.arange(1,15)
x1=2*x
y1=4*x
plt.plot(x,x1,color='g',linestyle=':',linewidth=4)
plt.plot(x,y1,color='b',linestyle='-',linewidth=4)
plt.title('Title')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)
plt.show()
```

```python
x=np.arange(1,15)
x1=2*x
y1=4*x

plt.subplot(1,2,1)
plt.plot(x,x1,color='g',linestyle=':',linewidth=4)
plt.title('Title')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)

plt.subplot(1,2,2)
plt.plot(x,y1,color='b',linestyle='-',linewidth=4)
plt.title('Title')
plt.xlabel('x-label')
plt.ylabel('y-label')
plt.grid(True)

plt.show()
```
