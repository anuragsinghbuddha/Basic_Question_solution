## Bank issue a credit card to customer under scheme of master card.Based on past data bank has found that only 60% of all account pay on time  if sample of 7 account is taken as random,construct binomial distribution  probability of paying on time.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
n=7 #no. of trail
p=.6 # probability of success
```

```python
k=np.arange(0,8) # no. of possible chance of success
k
```

```python
binomial=stats.binom.pmf(k,n,p) # applying probability mass function = nCr (P)**r (1-P)**n-r
binomial
```

```python
plt.plot(k,binomial,"o-")
plt.title('Binomial:n=%i,p%.2f'%(n,p),fontsize=15)
plt.xlabel("Number of success")
plt.ylabel("Probability of success")
plt.show()
```
